### Get Data and Create Chunks

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_openai.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
def get_data(url):
    # Retireve text from the document / url
    loader = WebBaseLoader(url)
    document = loader.load()
    documents_chunks = text_splitter.split_documents(document)
    # Split text into chunks
    # print(document)
    # Create a vector store
    vector_store = Chroma.from_documents(documents_chunks, embedding=embeddings)
    return documents_chunks
```

```python
url = "https://en.wikipedia.org/wiki/History_of_Wikipedia"
result = get_data(url)
result
```
